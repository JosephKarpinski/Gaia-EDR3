```python
BossQSO  = pd.read_csv("./Downloads/BossQSO_L64000.csv")

starclass = BossQSO

print("BOSS QSO SurveySize: " + str(starclass.shape[0]) + "\n")

subclass = sp.unique(starclass.subClass)

for s in subclass:
    print(str(s) + ": " + str(starclass[starclass.subClass == s]['subClass'].count()))
```

```python

df = BossQSO[BossQSO.subClass == "AGN"][0:10]


prefix = "AGN"


file_records = len(df)

for x in range(0, file_records):
    plate = df.iloc[x]['plate']
    mjd = df.iloc[x]['mjd']
    fiberID = df.iloc[x]['fiberID']
    
    if len(str(fiberID)) == 1:
        fiberLen = "000"
    else:
        if len(str(fiberID)) == 2:
            fiberLen = "00"
        else:
            if len(str(fiberID)) == 3:
                fiberLen = "0"
            else:
                fiberLen = ""
    try:
        urllib.request.urlretrieve("https://data.sdss.org/sas/dr12/boss/spectro/redux/v5_7_0/spectra/" + 
            str(plate) + "/" + "spec-" + str(plate) + "-" + str(mjd) + "-" + fiberLen + str(fiberID) + ".fits", 
            "./Downloads/BOSS_eBOSS FITS/" + str(prefix) + 
                  "_spec-" + str(plate) + "-" + str(mjd) + "-" + fiberLen + str(fiberID) + ".fits")
    except:
        print("Download Failed: " + "spec-" + str(plate) + "-" + str(mjd) + "-" + fiberLen + str(fiberID) + ".fits")
    
print("Done!")
```
